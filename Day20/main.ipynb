```python
with open('./input.txt') as f:
    data = f.read().split('\n\n') 
    enhancement = [x for x in data[0]]
    enhancement = [1 if x == '#' else 0 for x in enhancement]

    image = [ [1 if x=='#' else 0 for x  in row] for row in data[1].split('\n')][:-1]

enhancement

```

```python
def pad(img, value = 0):
    side = [value] *3
    #button and top pad
    top = [[value] * (len(img[0]) + 6)] * 3
    image = [*top, 
            *[[*side, *line, *side] for line in img],
             *top]
    return image
```

```python
def enhance(image, pad_value = 0):
    img = pad(image,pad_value)
    height, width = len(img),len(img[0])
    res = [[0] * width for _ in img]
    for i in range(height -2):
        for j in range(width -2):
            bits = [img[i1][j1] for i1 in range(i, i + 3) for j1 in range(j, j + 3)]
            index = int(''.join(map(str, bits)), 2)
            res[i + 1][j + 1] = enhancement[index]
    return [line[2:-2] for line in res[2:-2]]
```

```python
enhanced = enhance(enhance(image), pad_value=1)
print(sum(char for line in enhanced for char in line))
```

```python
enhanced = image
for i in range(50):
    enhanced = enhance(enhanced, i %2)
print(sum(char for line in enhanced for char in line))
```
